# Movie System Recommendation Using Collaborative Filtering

```python
import pandas as pd
import numpy as np 
import seaborn as sns
```

```python
movies = pd.read_csv('movies.csv')
```

```python
ratings = pd.read_csv('ratings.csv')
```

```python
movies
```

```python
df = pd.merge(movies, ratings, on='movieId')
```

```python
df
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean()).reset_index()
```

```python
df.groupby('title')['rating'].count().values
```

```python
ratings['num of ratings'] = df.groupby('title')['rating'].count().values
```

```python
ratings
```

```python
movie = df.pivot_table(index='userId', columns= 'title', values='rating')
```

```python
movie.head()
```

```python
movie = movie.fillna(0)
movie.head()
```

```python
similarity = movie.corr(method='pearson')
```

```python
similarity.head(10)
```

```python
def get_similar_movies(movie_name,user_rating) :
    score = similarity[movie_name]*(user_rating-2.5)
    score = score.sort_values(ascending=False)
    return score
```

```python
user1 = [('Toy Story (1995)',1),('Flint (2017)',1),('No Game No Life: Zero (2017)',1),('Titanic (1997)',1)]
```

```python
user2 = [('Toy Story (1995)',1),('Star Wars: Episode VII - The Force Awakens (2015)',5),('No Game No Life: Zero (2017)',4)]
```

```python
rec = pd.DataFrame()
for movie,rating in user2 :
    rec = rec.append(get_similar_movies(movie,rating),ignore_index=True)
    
```

```python
rec = rec.sum().sort_values(ascending=False)
```
