```python
unique_list = []
count_list = []
def unique(list1):
    # traverse for all elements
    for x in list1:   
        for y in x :
            if y not in unique_list:
                unique_list.append(y)
                count_list.append(1)
            else :
                count_list[unique_list.index(y)] += 1
            
unique(movie['genres'])
print(unique_list,count_list)
```
